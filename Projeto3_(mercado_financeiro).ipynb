```python
# Bibliotecas para Modelagem e Matrizes
import numpy as np
import pandas as pd

# Bibliotecas para análises gráficas
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go

# Bibliotaca para ignorar avisos
import warnings

# Desabilitando avisos
warnings.filterwarnings('ignore')
```

```python
# Leitura dos Dados
Base_Dados = pd.read_excel('Ações - Magalu - Sem Resolução.xlsx')

# Cabeçalho (primeiras 5 linhas da base de Dados)
Base_Dados.head()
```

```python
Base_Dados.shape
```

```python
Base_Dados.info()
```

```python
Base_Dados.describe()
```

```python
# Séries Temporais
Dados = Base_Dados.set_index('Data') #Coloca 'Data' como primeira coluna da Base de Dados
Dados.head()
```

```python
Base_Dados.tail() #últimas 5 linhas da base de Dados
```

```python
# Plotagem dos valores de fechamento por período de cotação
plt.style.use('seaborn-darkgrid')
plt.figure( figsize=(16, 5) )
plt.title('Análise das ações da Magalu - Fechamento', fontsize=15, loc='left')
plt.plot( Dados.index, Dados['Fechamento'] )

plt.xlabel('Período da Cotação')
plt.ylabel('Valor da Ação (R$)');
```

```python
Media_Movel = Dados['Fechamento'].rolling(5).mean()
Media_Tendencia = Dados['Fechamento'].rolling(30).mean()

plt.style.use('seaborn-darkgrid')
plt.figure( figsize=(16, 5) )
plt.title('Análise das ações da Magalu - Fechamento', fontsize=15, loc='left')

plt.plot( Dados.index, Dados['Fechamento'] )
plt.plot( Dados.index, Media_Movel )
plt.plot( Dados.index, Media_Tendencia )

plt.xlabel('Período da Cotação')
plt.ylabel('Valor da Ação (R$)');

plt.legend(["Fechamento","Média Móvel", "Tendência"])
```

```python
sns.boxplot( data=Dados, x='Fechamento')
```

```python
# Boxplot Mensal

Base_Dados['Mes'] = Base_Dados['Data'].dt.month

plt.figure( figsize=(16, 5) )
sns.boxplot( data=Base_Dados, x='Mes', y='Fechamento' );
```

```python
Base_Dados.groupby(['Mes']).describe()['Fechamento']
```

```python
# Gráfico de candlestick
Grafico = go.Figure(
    data=[
          go.Candlestick(
              x= Dados.index,
              open = Dados['Abertura'],
              high = Dados['Maior'],
              low = Dados['Menor'],
              close = Dados['Fechamento'],
          )
    ]
)

Grafico.show()
```

```python
# Gráfico de candlestick
Grafico.update_layout( xaxis_rangeslider_visible=False ) #remove janela deslizante abaixo do gráfico

Grafico.show()
```
